```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/mcdonalds.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df,x='Age')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Age distribution')
plt.show()
```

```python
to_drop=df.columns[11:] #dropped columns will be used as profilers
df_tocluster=df.drop(to_drop,axis=1)
df_tocluster.head()
```

```python
le=LabelEncoder()
label=df_tocluster.columns[:11]
for col in label:
  df_tocluster[col]=le.fit_transform(df_tocluster[col])
df_tocluster.head()
```

```python
#Here Standard Scaler is not needed since we are considering only categorical variables in the clustring process
```

```python
#elbow method for finding optimum no. of clusters
wcss=[]
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(df_tocluster)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS(Inertia)')
plt.title('Elbow Method')
```

```python
from sklearn.metrics import silhouette_score
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,random_state=42)
  labels=kmeans.fit_predict(df_tocluster)
  score=silhouette_score(df_tocluster,labels)
  print(f'Silhouette score for K= {k} : {score:.3f}')
```

```python
kmeans=KMeans(n_clusters=4, random_state=42) #selecting 4 clusters and clustering
df['Cluster']= kmeans.fit_predict(df_tocluster)
df
```

```python
df['Cluster']= kmeans.labels_
```

```python
features= ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']
X=df_tocluster[features]

#apply PCA
pca=PCA(n_components=2)
pca_result= pca.fit_transform(X)
df['PCA1']=pca_result[:,0]
df['PCA2']=pca_result[:,1]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='PCA1',y='PCA2',hue='Cluster',palette='Set2',alpha=0.7)
plt.title('PCA Visualization')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Mosaic equivalent (stacked bar for each segment)
sns.countplot(data=df, x='Cluster', hue='Like')
plt.title("Like Distribution by Cluster")
plt.show()

sns.countplot(data=df, x='Cluster', hue='Gender')
plt.title("Gender Distribution by Cluster")
plt.show()

sns.countplot(data=df, x='Cluster', hue='Age')
plt.title("Age Distribution by Cluster")
plt.show()

sns.countplot(data=df, x='Cluster', hue='VisitFrequency')
plt.title("VisitFrequency Distribution by Cluster")
plt.show()
```

```python
# Convert Like to numerical representation
df['Like_Num'] = df['Like'].map({'I hate it!-5': -5, 'I love it!+5': 5}) # Assume -5 to 5 scale based on the error


# Convert VisitFrequency to numerical representation
visit_frequency_mapping = {
    'Never': 0,
    'Once a year': 1,
    'Every three months': 2,
    'Once a month': 3,
    'More than once a week':4,
    'Once a week':5
}
df['VisitFrequency_Num'] = df['VisitFrequency'].map(visit_frequency_mapping)

# Create 'Female' column (1 for Female, 0 for Male)
df['Female'] = (df['Gender'] == 'Female').astype(int)

# Now you can proceed with the groupby and agg operation
segment_summary = df.groupby('Cluster').agg({
    'Like_Num': 'mean',
    'VisitFrequency_Num': 'mean',
    'Female': 'mean'
}).reset_index()
```

```python
#Bubble evaluation chart for Like and VisitFrquency
segment_summary = df.groupby('Cluster').agg({
    'Like_Num': 'mean',
    'VisitFrequency_Num': 'mean',
    'Female': 'mean'
}).reset_index()

plt.figure(figsize=(8,6))
plt.scatter(segment_summary['VisitFrequency_Num'],
            segment_summary['Like_Num'],
            s=segment_summary['Female'] * 500 + 100,
            alpha=0.7, edgecolors='black')

for i in range(len(segment_summary)):
    plt.text(segment_summary['VisitFrequency_Num'][i] + 0.01,
             segment_summary['Like_Num'][i] + 0.01,
             f"Seg {segment_summary['Cluster'][i]}", fontsize=10)

plt.xlabel("Visit Frequency (mean)")
plt.ylabel("Like (mean)")
plt.title("Segment Evaluation Plot")
plt.grid(True)
plt.show()
```

```python
# Calculate means of perception variables by cluster
perception_vars = df.columns[:11]  # first 11 columns (Yummy, Tasty, etc.)
cluster_perception_means = df.groupby('Cluster')[perception_vars].apply(lambda x: (x == "Yes").mean())

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(cluster_perception_means, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Perception Variable Mean Scores by Cluster")
plt.xlabel("Perception Variables")
plt.ylabel("Cluster")
plt.tight_layout()
plt.show()
```
